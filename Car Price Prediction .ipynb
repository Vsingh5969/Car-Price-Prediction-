# Training model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
```

#### Find the best random state;

```python
max_accu=0
max_RS=0

for i in range(0,500):
    x_train,x_test,y_train,y_test= train_test_split(x_scaler,y,random_state=i,test_size=0.35)
    lr= LinearRegression()
    lr.fit(x_train,y_train)
    pred= lr.predict(x_test)
    r2score=r2_score(y_test,pred)
    if r2score>max_accu:
        max_accu=r2score
        max_RS=i
print("Best accuracy is ",max_accu,"on Random State =",max_RS)
```

```python
 x_train,x_test,y_train,y_test= train_test_split(x_scaler,y,random_state=111,test_size=0.20)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.linear_model import SGDRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.tree import ExtraTreeRegressor
from sklearn.neural_network import MLPRegressor
```

```python
lr_model=LinearRegression()
dt_model=DecisionTreeRegressor()
svc_model=SVR()
knn_model=KNeighborsRegressor()
rd_model=RandomForestRegressor()
xgb_model=XGBRegressor()
sdg_model=SGDRegressor()
bag_model=BaggingRegressor()
ada_model=AdaBoostRegressor()
gd_model=GradientBoostingRegressor()
et_model=ExtraTreeRegressor()
nural_model=MLPRegressor()
```

```python
models=[lr_model,dt_model,svc_model,knn_model,rd_model,xgb_model,sdg_model,bag_model,ada_model,gd_model,et_model,nural_model]
```

```python
for m in models:
    m.fit(x_train,y_train)
    m.score(x_train,y_train)
    predict=m.predict(x_test)
    print("Adjusted R2 squared of ",m,'is:',m.score(x_train,y_train))
    print("mean_absolute_error", mean_absolute_error(y_test,predict))
    print('mean_squared _error ',m,' is ',mean_squared_error(y_test,predict))
    print('Root mean Squared error',np.sqrt(mean_squared_error(y_test,predict)))
    print('*******************************************')
```
